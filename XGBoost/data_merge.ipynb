## **df_mst 가장 근접한 역가지의 거리, 역사명, 호선명 추가**

```python
import numpy as np
from scipy.spatial import cKDTree
from tqdm import notebook, trange, tqdm

def haversine(lat1, lon1, lat2, lon2): # 위도, 경도를 이용하여 거리 계산 함수
    # 지구 반지름 (km)
    R = 6371.0
    
    # 위도와 경도를 라디안 단위로 변환
    phi1 = np.radians(lat1)
    phi2 = np.radians(lat2)
    delta_phi = np.radians(lat2 - lat1)
    delta_lambda = np.radians(lon2 - lon1)
    
    # 하버사인 공식 계산
    a = np.sin(delta_phi / 2.0)**2 + np.cos(phi1) * np.cos(phi2) * np.sin(delta_lambda / 2.0)**2
    c = 2 * np.arctan2(np.sqrt(a), np.sqrt(1 - a))
    
    # 거리 계산
    distance = R * c
    return distance

def find_near_place(df1, df2):

    om_tree= cKDTree(df2[['위도', '경도']].values) # 대기질 데이터의 위도, 경도를 이용하여 cKDTree 생성
    distance, index = om_tree.query(df1[['위도', '경도']].values, k=1)
    actual_distance = np.array([haversine(df1.iloc[i]['위도'], df1.iloc[i]['경도'], df2.iloc[index[i]]['위도'], df2.iloc[index[i]]['경도']) for i in range(len(df1))]) # 거리 계산 함수를 이용하여 거리 계산 

    return actual_distance, index


```

## **df_anc 가장 근접한 역가지의 거리, 역사명, 호선명 추가**

```python
actual_distance, index = find_near_place(df_anc, df_subway)

df_anc['지하철역까지_거리'] = actual_distance
df_anc['지하철역명'] = df_subway.loc[index, '역사명'].reset_index(drop=True)
df_anc['호선'] = df_subway.loc[index, '호선'].reset_index(drop=True)
```
