```python
import pandas as pd
from tqdm import tqdm
import requests
import numpy as np
```

```python
df_env = pd.read_feather("../data/make_data/생활환경지수.feather")
```

```python
# 구글에서 제공하는 API를 사용하여 주소를 기반으로 위도와 경도를 추가한다.


# df_soc

import requests

api_key = "" # 구글 API 키


address_coordinates = {} # 주소를 기반으로 위도, 경도 데이터를 저장할 딕셔너리
for address in enumerate(tqdm(df_env['지번주소'])):

    url = f"https://maps.googleapis.com/maps/api/geocode/json?address={address}&key={api_key}" # 구글 API를 이용하여 위도, 경도 데이터를 가져옴
    
    response = requests.get(url) # get 요청
    
    data = response.json() # json 데이터로 변환
    
    if data['status'] == 'OK': # 정상적으로 데이터를 가져왔을 때
        lat = data['results'][0]['geometry']['location']['lat']
        lng = data['results'][0]['geometry']['location']['lng']
        address_coordinates[address] = (lat, lng) # 위도, 경도 데이터 추가

for address, coordinates in address_coordinates.items():
    print(f"Address: {address}")
    print(f"Latitude: {coordinates[0]}")
    print(f"Longitude: {coordinates[1]}")
    print()
```

```python
# 주소를 기반으로 위도와 경도를 추가
df_env['위도'] = np.nan
df_env['경도'] = np.nan

for address, coordinates in tqdm(address_coordinates.items()):
    df_env.loc[df_env['지번주소']==address[1], '위도']= coordinates[0]
    df_env.loc[df_env['지번주소']==address[1], '경도']= coordinates[1]
```

```python
df_env.to_feather("../data/make_data/df_env.feather")
```
